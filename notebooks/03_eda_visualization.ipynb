```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/cleaned/social_media_cleaned.csv")
df.head()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="daily_social_media_time", y="actual_productivity_score", data=df)

plt.title("Social Media Time vs Productivity")
plt.xlabel("Daily Social Media Time")
plt.ylabel("Actual Productivity Score")

plt.show()
```

Higher social media usage tends to slightly reduce productivity scores.



```python
plt.figure(figsize=(8,5))
sns.boxplot(x="stress_level", y="actual_productivity_score", data=df)

plt.title("Stress Level vs Productivity")

plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="sleep_hours", y="actual_productivity_score", data=df)

plt.title("Sleep Hours vs Productivity")

plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="social_platform_preference", data=df)

plt.title("Most Used Social Media Platforms")

plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="number_of_notifications", y="stress_level", data=df)

plt.title("Notifications vs Stress Level")

plt.show()
```

```python
plt.figure(figsize=(12,8))

sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")

plt.title("Feature Correlation Heatmap")

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.boxplot(x="job_type", y="actual_productivity_score", data=df)

plt.xticks(rotation=45)

plt.title("Productivity Score by Job Type")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["daily_social_media_time"], bins=30, kde=True)

plt.title("Distribution of Daily Social Media Usage")

plt.show()
```

```python
plt.savefig("../images/social_media_vs_productivity.png")
```

EDA Summary

Exploratory analysis revealed several key insights.

Higher social media usage tends to slightly reduce productivity levels.

Stress level has a negative impact on productivity.

Adequate sleep hours appear to improve productivity.

Frequent notifications may contribute to higher stress levels.

Different job types show varying productivity patterns.

These insights will be useful for building machine learning models to predict productivity.

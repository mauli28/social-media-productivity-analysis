```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/raw/social_media_vs_productivity.csv")

df.head()
```

```python
df.shape
```

Dataset contains 30,000 rows and 19 columns.



```python
df.isnull().sum()
```

```python
df = df.dropna(subset=["actual_productivity_score"])
```

```python
df["daily_social_media_time"].fillna(df["daily_social_media_time"].mean(), inplace=True)

df["perceived_productivity_score"].fillna(df["perceived_productivity_score"].mean(), inplace=True)

df["stress_level"].fillna(df["stress_level"].mean(), inplace=True)

df["sleep_hours"].fillna(df["sleep_hours"].mean(), inplace=True)

df["screen_time_before_sleep"].fillna(df["screen_time_before_sleep"].mean(), inplace=True)

df["job_satisfaction_score"].fillna(df["job_satisfaction_score"].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.dtypes
```

```python
df.to_csv("../data/cleaned/social_media_cleaned.csv", index=False)
```

Data Cleaning Summary

Missing values were detected in several behavioral and lifestyle columns.

Rows with missing target variable (actual_productivity_score) were removed.

Remaining missing values were handled using mean imputation.

Dataset was checked for duplicate rows.

Cleaned dataset saved for further analysis.

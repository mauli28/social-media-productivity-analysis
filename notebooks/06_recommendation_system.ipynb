```python
import pandas as pd
```

```python
df = pd.read_csv("../data/cleaned/social_media_scored.csv")

df.head()
```

```python
def generate_recommendations(row):
    
    recommendations = []
    
    if row["daily_social_media_time"] > 4:
        recommendations.append("Reduce daily social media usage")
        
    if row["sleep_hours"] < 6:
        recommendations.append("Increase sleep hours for better focus")
        
    if row["stress_level"] > 7:
        recommendations.append("Manage stress using relaxation techniques")
        
    if row["screen_time_before_sleep"] > 2:
        recommendations.append("Avoid screen time before sleep")
        
    if row["breaks_during_work"] < 2:
        recommendations.append("Take short breaks during work for productivity")
        
    if row["uses_focus_apps"] == False:
        recommendations.append("Consider using focus apps to avoid distractions")
        
    return recommendations
```

```python
df["recommendations"] = df.apply(generate_recommendations, axis=1)
```

```python
df[[
    "productivity_score",
    "productivity_level",
    "recommendations"
]].head()
```

```python
df["productivity_score"] = df["productivity_score"].fillna(df["productivity_score"].mean())
```

```python
df["productivity_score"].isnull().sum()
```

```python
df.to_csv("../data/cleaned/social_media_final.csv", index=False)
```
